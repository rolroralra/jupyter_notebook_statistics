```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
df = pd.read_csv("../data/bike.csv")
df.head(2)
```

```python
df_sub = df.iloc[:, [1, 2, 3, 4, 5, 6]]
df_sub.head(2)
```

```python
df_sub.corr().style.background_gradient(cmap = "PuBu")
```

```python
df_sub.values

import numpy as np
import pandas as pd

df_sub2 = pd.concat([pd.DataFrame(np.ones_like(df_sub.iloc[:,0]), columns=["Intercept"]), df_sub], axis=1)
df_sub2
```

```python
df_vif = pd.DataFrame()
df_vif["VIF"] = [vif(df_sub2.values, i) for i in range(df_sub2.shape[1])]
df_vif["features"] = df_sub2.columns
df_vif
```

```python
df_vif = pd.DataFrame()
df_vif["VIF"] = [vif(df_sub.values, i) for i in range(df_sub.shape[1])]
df_vif["features"] = df_sub.columns
df_vif
```
